## 구슬 탈출 2(13460번)

```python
from collections import deque

n,m=map(int,input().split())

board = [list(input().rstrip()) for _ in range(n)] 

queue=[]

dx=[-1,1,0,0]
dy=[0,0,-1,1]

visited=[[[[False]*m for _ in range(n)]for _ in range(m)]for _ in range(n)]

def init():
    rx,ry,bx,by=0,0,0,0
    for i in range(n):
        for j in range(m):
            if board[i][j]=='R':
                rx,ry=i,j
            elif board[i][j]=='B':
                bx,by=i,j
    queue.append((rx,ry,bx,by,1))
    visited[rx][ry][bx][by]=True
    
def move(x,y,dx,dy):
    cnt=0
    while board[x+dx][y+dy]!='#' and board[x][y]!='O':
        x+=dx
        y+=dy
        cnt+=1
    return x,y,cnt

def solve():
    init()
    while queue:
        rx,ry,bx,by,depth=queue.pop(0)
        print(rx,ry,bx,by)
        if depth>10:
            break
        for i in range(4):
            nrx,nry,rcnt=move(rx,ry,dx[i],dy[i])
            print('red : ',i,nrx,nry,rcnt)
            nbx,nby,bcnt=move(bx,by,dx[i],dy[i])
            print('blue : ',i,nrx,nry,rcnt)
            if board[nbx][nby]!='O':
                if board[nrx][nry]=='O':
                    print(depth)
                    return
                if nrx==nbx and nry==nby:
                    if rcnt>bcnt:
                        nrx-=dx[i]
                        nry-=dy[i]
                    else:
                        nbx-=dx[i]
                        nby-=dy[i]
                if not visited[nrx][nry][nbx][nby]:
                    visited[nrx][nry][nbx][nby]=True
                    queue.append((nrx,nry,nbx,nby,depth+1))
    print(-1)
    
solve()
```

```python
from collections import deque

n,m=map(int,input().split())
#0:북  1: 동  2:남  3:서 
cur_x,cur_y,cur_dir=map(int,input().split())

graph=[]
for i in range(n):
    graph.append(list(map(int,input().split())))
#방문 처리    
visited=[[False]*m for _ in range(n)]

#이동 방향
#0:북  1: 동  2:남  3:서 
dx=[-1,0,1,0]
dy=[0,1,0,-1]

def turn_left(d):
    result=(d-1)%4
    return result

def bfs():
    global clean
    q=deque()
    q.append([cur_x,cur_y,cur_dir])
    visited[cur_x][cur_y]=True
    while q:
        x,y,d=q.popleft()
        #이동하여 청소 여부 확인
        is_possible=False
        for i in range(4):
            d=turn_left(d)
            nx=x+dx[d]
            ny=y+dy[d]
            
            if 0<=nx<n and 0<=ny<m and graph[nx][ny]==0:
                #청소가 안되있는 구역이라면 청소
                if visited[nx][ny]==False:
                    #이동 완료
                    visited[nx][ny]=True
                    clean+=1
                    q.append([nx,ny,d])
                    #이동하여 청소 여부 확인
                    is_possible=True
                    break
        
        # 이동해서 청소할 수 있는 영역이 없는경우
        if is_possible==False:
            #뒤로 후진
            nx=x-dx[d]
            ny=y-dy[d]
            if graph[nx][ny]==1:
                break
            else:
                q.append([nx,ny,d])

clean=1
bfs()
print(clean)
```

```python
n,m=map(int,input().split())
graph=[]
for _ in range(n):
    data=list(map(int,input().split()))
    graph.append(data)
    
dx=[1,-1,0,0]
dy=[0,0,1,-1]

#그래프 중 가장 큰값
maxv = max(map(max, graph))

def dfs(x,y,depth,s):
    global max_value
    
    #갈 수 있는 남은 곳이 최대값보다 큰 경우 조기 종료
    if s+maxv*(4-depth)<=max_value:
        return
    if depth>=4:
        if max_value<s:
            max_value=s
        return
    else:
        for i in range(4):
            nx=x+dx[i]
            ny=y+dy[i]
            if 0<=nx<n and 0<=ny<m and visited[nx][ny]==False:
                # ㅗ모양을 해결하기 2칸일 때 옆이 아닌  칸으로 이동
                if depth==2:
                    visited[nx][ny]=True
                    dfs(x,y,depth+1,s+graph[nx][ny])
                    visited[nx][ny]=False
                visited[nx][ny]=True
                dfs(nx,ny,depth+1,s+graph[nx][ny])
                #갔다온 곳 풀어주기
                visited[nx][ny]=False
                
max_value=0
visited = [[False]*m for _ in range(n)]
for i in range(n):
    for j in range(m):
        visited[i][j] = True
        dfs(i,j,1,graph[i][j])
        visited[i][j] = False
        
print(max_value)
```

# 스타트와 링크(14889)
https://www.acmicpc.net/problem/14889

```python
#조합 이용
from itertools import combinations

n=int(input())
graph=[]
# 팀 멤버 구성
team=[]
for i in range(n):
    team.append(i)
    data=list(map(int,input().split()))
    graph.append(data)
# 전체 인원 중 반으로 팀을 나눈다
team_num=int(n/2)

# 반 중에 가능한 경우의 수
candidates=list(combinations(team,team_num))

result=1e9
for candidate in candidates:
    tmp=[]
    # 상대 팀의 배열
    for i in team:
        if i not in candidate:
            tmp.append(i)
    #각 팀의 점수를 구하기 위해 nC2로 묶기
    team1=list(combinations(candidate,2))
    team2=list(combinations(tmp,2))
    team1_score,team2_score=0,0
    for x,y in team1:
        team1_score+=graph[x][y]+graph[y][x]
    for x,y in team2:
        team2_score+=graph[x][y]+graph[y][x]
    # 최소값인 경우를 찾기 위해
    result=min(result,abs(team1_score-team2_score))
print(result)
```

```python
#depth는 팀 수를 의미,
def dfs(depth, idx):
    global min_diff
    # 팀원이 n//2명이 채워지면 점수를 내고 끝내기
    if depth == n//2:
        power1, power2 = 0, 0
        for i in range(n):
            for j in range(n):
                if visited[i] and visited[j]:
                    power1 += graph[i][j]
                elif not visited[i] and not visited[j]:
                    power2 += graph[i][j]
        min_diff = min(min_diff, abs(power1-power2))
        return
    
    # 팀원 선택하기, 
    for i in range(idx, n):
        if not visited[i]:
            # 팀원 선택
            visited[i] = True
            # 백트래킹을 통해 모든 경우의 수를 수행해 min_diff의 가장 작은값을 도출
            dfs(depth+1, i+1)
            # 팀원 선택한 것을 제거
            visited[i] = False


n = int(input())

visited = [False for _ in range(n)]
graph = [list(map(int, input().split())) for _ in range(n)]
min_diff = int(1e9)

dfs(0, 0)
print(min_diff)
```

```python
n,m,cx,cy,k=map(int,input().split())
graph=[]
for i in range(n):
    data=list(map(int,input().split()))
    graph.append(data)
    
cmd=list(map(int,input().split()))

# 주사위 :0번 index가 윗면인 경우,5번 index가 아랫면 인경우
# 주사위 초기상태 모두 0
dice=[0,0,0,0,0,0]


# 동: 1 서: 2 북: 3 남: 4
dx=[0,0,-1,1]
dy=[1,-1,0,0]
    
def rotate(direct):
    a,b,c,d,e,f=dice[0],dice[1],dice[2],dice[3],dice[4],dice[5]
    #동쪽으로 회전
    if direct==1:
        dice[0],dice[1],dice[2],dice[3],dice[4],dice[5]=d,b,a,f,e,c
    #서
    elif direct==2:
        dice[0],dice[1],dice[2],dice[3],dice[4],dice[5]=c,b,f,a,e,d
    #북
    elif direct==3:
        dice[0],dice[1],dice[2],dice[3],dice[4],dice[5]=e,a,c,d,f,b
    #남
    else:
        dice[0],dice[1],dice[2],dice[3],dice[4],dice[5]=b,f,c,d,a,e   

nx,ny=cx,cy
for direct in cmd:
    # 방향에 맞춰 주사위 굴리기
    nx=nx+dx[direct-1]
    ny=ny+dy[direct-1]
#     if 0<=nx<n and 0<=ny<m:
    # 범위를 넘어 가지 못하는 경우 제자리 위치
    if 0>nx or nx>=n or ny>=m or ny<0:
        nx-=dx[direct-1]
        ny-=dy[direct-1]
        continue
    
    rotate(direct)

    #주사위 숫자 회전
    # 보드가 0인경우 주사위 아랫면 복사
    if graph[nx][ny]==0:
        graph[nx][ny]=dice[-1]
    # 보드가 0이 아니면 보드 값을 주사위에 복사 후 보드 0으로 바꾸기
    else:
        dice[-1]=graph[nx][ny]
        graph[nx][ny]=0

    print(dice[0])
```
