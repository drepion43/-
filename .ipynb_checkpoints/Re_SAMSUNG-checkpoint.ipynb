# 2048
https://www.acmicpc.net/problem/12100

```python
import copy

n = int(input())
board = []
for _ in range(n):
    data = list(map(int,input().split()))
    board.append(data)

# 위료
def up(board):
    for y in range(n):
        #제일 꼭대기
        start = 0
        for x in range(1, n):
            if board[x][y]:
                # 아래 데이터 위로 이동
                data = board[x][y]
                board[x][y] = 0
                #위가 비어있는 경우
                if board[start][y] == 0:
                    board[start][y] = data
                #아래와 같은 경우
                elif board[start][y] == data:
                    board[start][y] *= 2
                    start += 1
                # 다른 경우
                else:
                    start +=1
                    board[start][y] = data
    return board

# 아래로

def down(board):
    for y in range(n):
        #제일 아래
        start = n-1
        for x in range(n-2, -1, -1):
            if board[x][y]:
                # 위 데이터 아래로 이동
                data = board[x][y]
                board[x][y] = 0
                #아래가 비어있는 경우
                if board[start][y] == 0:
                    board[start][y] = data
                #아래와 같은 경우
                elif board[start][y] == data:
                    board[start][y] *= 2
                    start -= 1
                # 다른 경우
                else:
                    start -=1
                    board[start][y] = data
    return board

# 우측
def right(board):
    for x in range(n):
        #제일 우측
        start = n-1
        for y in range(n-2,-1,-1):
            if board[x][y]:
                # 좌측 데이터 우측으로 이동
                data = board[x][y]
                board[x][y] = 0
                #우측이 비어있는 경우
                if board[x][start] == 0:
                    board[x][start] = data
                #좌측과 같은 경우
                elif board[x][start] == data:
                    board[x][start] *= 2
                    start -= 1
                # 다른 경우
                else:
                    start -=1
                    board[x][start] = data
    return board

# 좌측
def left(board):
    for x in range(n):
        #제일 좌측
        start = 0
        for y in range(1, n):
            if board[x][y]:
                # 우측 데이터 좌측으로 이동
                data = board[x][y]
                board[x][y] = 0
                # 좌측이 비어있는 경우
                if board[x][start] == 0:
                    board[x][start] = data
                # 우측과 같은 경우
                elif board[x][start] == data:
                    board[x][start] *= 2
                    start += 1
                # 다른 경우
                else:
                    start +=1
                    board[x][start] = data
    return board

def dfs(board, cnt):
    if cnt==5:
        return max(map(max, board))
    return max(dfs(up(copy.deepcopy(board)),cnt+1), dfs(down(copy.deepcopy(board)),cnt+1), dfs(left(copy.deepcopy(board)),cnt+1), dfs(right(copy.deepcopy(board)),cnt+1))

print(dfs(board,0))
```

# 뱀
https://www.acmicpc.net/problem/3190

```python
from collections import deque
n = int(input())
k = int(input())

#사과 위치
apples = []
for _ in range(k):
    x,y = map(int, input().split())
    apples.append([x-1,y-1])

l = int(input())

# 뱀 이동
snake_dirs = deque()
acc_x = 0
for _ in range(l):
    x, c = input().split()
    acc_x += int(x)
    snake_dirs.append([int(x),c])

# 보드 생성
# 뱀 : 1, 사과 : 2 빈곳 : 0
board = [[0] * n for _ in range(n)]
for x,y in apples:
    board[x][y] = 2



# 우측 90도 회전 +1 , 좌측 90도 이동 -1
dx = [0,1,0,-1]
dy = [1,0,-1,0]
directs = {'D' : 1, 'L':-1}
#뱀 회전
def turn(d,cd):
    if d=='L':
        return (cd-1) %4
    else:
        return (cd+1) %4

#초기화 정의
x,y = 0,0
board[x][y] = 1
cnt = 0
d = 0
snake_idx = 0
#뱀 위치
q = deque()
q.append([0,0])

while True:
    cnt+=1
    x, y = x + dx[d], y + dy[d]
    # print(board, snake_dirs[snake_idx][0], d, cnt)
    if 0<=x<n and 0<=y<n:
        #이동위치에 사과가 있을시
        if board[x][y]==2:
            #1은 뱀 있는 공간
            board[x][y] = 1
            q.append([x,y])
            if cnt == snake_dirs[snake_idx][0]:
                d = turn(snake_dirs[snake_idx][1],d)
                if snake_idx < l-1:
                    snake_idx+=1
        # 사과 없을시
        elif board[x][y]==0:
            board[x][y] = 1
            q.append([x,y])
            cx,cy = q.popleft()
            board[cx][cy] = 0
            if cnt == snake_dirs[snake_idx][0]:
                d = turn(snake_dirs[snake_idx][1], d)
                if snake_idx < l-1:
                    snake_idx += 1
        else:
            break
    else:
        break


print(cnt)
```

# 시험감독
https://www.acmicpc.net/problem/13458

```python
n = int(input())
tester = list(map(int,input().split()))
b, c = map(int,input().split())

result = 0

for i in range(n):
    #총감독관
    if tester[i]>0:
        tester[i] -= b
        result += 1

    #부감독관
    if tester[i]>0:
        result += tester[i] // c
        if tester[i] % c:
            result += 1
print(result)
```
