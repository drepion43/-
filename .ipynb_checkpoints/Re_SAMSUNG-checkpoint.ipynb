# 2048
https://www.acmicpc.net/problem/12100

```python
import copy

n = int(input())
board = []
for _ in range(n):
    data = list(map(int,input().split()))
    board.append(data)

# 위료
def up(board):
    for y in range(n):
        #제일 꼭대기
        start = 0
        for x in range(1, n):
            if board[x][y]:
                # 아래 데이터 위로 이동
                data = board[x][y]
                board[x][y] = 0
                #위가 비어있는 경우
                if board[start][y] == 0:
                    board[start][y] = data
                #아래와 같은 경우
                elif board[start][y] == data:
                    board[start][y] *= 2
                    start += 1
                # 다른 경우
                else:
                    start +=1
                    board[start][y] = data
    return board

# 아래로

def down(board):
    for y in range(n):
        #제일 아래
        start = n-1
        for x in range(n-2, -1, -1):
            if board[x][y]:
                # 위 데이터 아래로 이동
                data = board[x][y]
                board[x][y] = 0
                #아래가 비어있는 경우
                if board[start][y] == 0:
                    board[start][y] = data
                #아래와 같은 경우
                elif board[start][y] == data:
                    board[start][y] *= 2
                    start -= 1
                # 다른 경우
                else:
                    start -=1
                    board[start][y] = data
    return board

# 우측
def right(board):
    for x in range(n):
        #제일 우측
        start = n-1
        for y in range(n-2,-1,-1):
            if board[x][y]:
                # 좌측 데이터 우측으로 이동
                data = board[x][y]
                board[x][y] = 0
                #우측이 비어있는 경우
                if board[x][start] == 0:
                    board[x][start] = data
                #좌측과 같은 경우
                elif board[x][start] == data:
                    board[x][start] *= 2
                    start -= 1
                # 다른 경우
                else:
                    start -=1
                    board[x][start] = data
    return board

# 좌측
def left(board):
    for x in range(n):
        #제일 좌측
        start = 0
        for y in range(1, n):
            if board[x][y]:
                # 우측 데이터 좌측으로 이동
                data = board[x][y]
                board[x][y] = 0
                # 좌측이 비어있는 경우
                if board[x][start] == 0:
                    board[x][start] = data
                # 우측과 같은 경우
                elif board[x][start] == data:
                    board[x][start] *= 2
                    start += 1
                # 다른 경우
                else:
                    start +=1
                    board[x][start] = data
    return board

def dfs(board, cnt):
    if cnt==5:
        return max(map(max, board))
    return max(dfs(up(copy.deepcopy(board)),cnt+1), dfs(down(copy.deepcopy(board)),cnt+1), dfs(left(copy.deepcopy(board)),cnt+1), dfs(right(copy.deepcopy(board)),cnt+1))

print(dfs(board,0))
```

# 뱀
https://www.acmicpc.net/problem/3190

```python
from collections import deque
n = int(input())
k = int(input())

#사과 위치
apples = []
for _ in range(k):
    x,y = map(int, input().split())
    apples.append([x-1,y-1])

l = int(input())

# 뱀 이동
snake_dirs = deque()
acc_x = 0
for _ in range(l):
    x, c = input().split()
    acc_x += int(x)
    snake_dirs.append([int(x),c])

# 보드 생성
# 뱀 : 1, 사과 : 2 빈곳 : 0
board = [[0] * n for _ in range(n)]
for x,y in apples:
    board[x][y] = 2



# 우측 90도 회전 +1 , 좌측 90도 이동 -1
dx = [0,1,0,-1]
dy = [1,0,-1,0]
directs = {'D' : 1, 'L':-1}
#뱀 회전
def turn(d,cd):
    if d=='L':
        return (cd-1) %4
    else:
        return (cd+1) %4

#초기화 정의
x,y = 0,0
board[x][y] = 1
cnt = 0
d = 0
snake_idx = 0
#뱀 위치
q = deque()
q.append([0,0])

while True:
    cnt+=1
    x, y = x + dx[d], y + dy[d]
    # print(board, snake_dirs[snake_idx][0], d, cnt)
    if 0<=x<n and 0<=y<n:
        #이동위치에 사과가 있을시
        if board[x][y]==2:
            #1은 뱀 있는 공간
            board[x][y] = 1
            q.append([x,y])
            if cnt == snake_dirs[snake_idx][0]:
                d = turn(snake_dirs[snake_idx][1],d)
                if snake_idx < l-1:
                    snake_idx+=1
        # 사과 없을시
        elif board[x][y]==0:
            board[x][y] = 1
            q.append([x,y])
            cx,cy = q.popleft()
            board[cx][cy] = 0
            if cnt == snake_dirs[snake_idx][0]:
                d = turn(snake_dirs[snake_idx][1], d)
                if snake_idx < l-1:
                    snake_idx += 1
        else:
            break
    else:
        break


print(cnt)
```

# 시험감독
https://www.acmicpc.net/problem/13458

```python
n = int(input())
tester = list(map(int,input().split()))
b, c = map(int,input().split())

result = 0

for i in range(n):
    #총감독관
    if tester[i]>0:
        tester[i] -= b
        result += 1

    #부감독관
    if tester[i]>0:
        result += tester[i] // c
        if tester[i] % c:
            result += 1
print(result)
```

#  주사위 굴리기
https://www.acmicpc.net/problem/14499

```python
from collections import deque

n,m,x,y,k = map(int,input().split())

board = []

for _ in range(n):
    data = list(map(int,input().split()))
    board.append(data)

cmd = list(map(int,input().split()))

# a,b,c,d,e,f = 윗,좌,우,앞,뒤,밑
dice = [0,0,0,0,0,0]

# 1: 동, 2: 서, 북: 3, 남:4
dx = [0,0,-1,1]
dy = [1,-1,0,0]
answer = []

# 주사위 굴리기
def roll(direct):
    a,b,c,d,e,f = dice[0], dice[1], dice[2], dice[3], dice[4], dice[5]
    if direct == 1:
        dice[0], dice[1], dice[2], dice[3], dice[4], dice[5] = d,a,c,f,e,b
    elif direct == 2:
        dice[0], dice[1], dice[2], dice[3], dice[4], dice[5] = b,f,c,a,e,d
    elif direct == 3:
        dice[0], dice[1], dice[2], dice[3], dice[4], dice[5] = c,b,f,d,a,e
    else:
        dice[0], dice[1], dice[2], dice[3], dice[4], dice[5] = e,b,a,d,f,c

nx,ny = x,y
while cmd:
    c = cmd.pop(0)
    nx,ny = nx+dx[c-1],ny+dy[c-1]
    if nx < 0 or nx >= n or ny < 0 or ny >= m:
        nx -= dx[c-1]
        ny -= dy[c-1]
        continue
    roll(c)
    if board[nx][ny] == 0:
        board[nx][ny] = dice[-1]
    else:
        dice[-1] = board[nx][ny]
        board[nx][ny] = 0
    print(dice[0])
```

# 테트리미노
https://www.acmicpc.net/problem/14500

```python
n,m = map(int,input().split())
board = []
for _ in range(n):
    data = list(map(int,input().split()))
    board.append(data)

dx = [1,-1,0,0]
dy = [0,0,1,-1]

total_max = max(map(max,board))
answer = 0

def dfs(x,y,depth,point):
    global answer
    #조기 종료
    if point+total_max*(4-depth)<=answer:
        return

    if depth>=4:
        answer = max(answer, point)
        return

    for i in range(4):
        nx,ny = x+dx[i], y+dy[i]
        if 0<=nx<n and 0<=ny<m and not visited[nx][ny]:
            # ㅗ 모양 고려
            if depth==2:
                visited[nx][ny]=True
                dfs(x,y,depth+1,point+board[nx][ny])
                visited[nx][ny]=False
            visited[nx][ny] = True
            dfs(nx,ny,depth+1,point+board[nx][ny])
            visited[nx][ny]=False

visited = [[False]*m for _ in range(n)]

for i in range(n):
    for j in range(m):
        visited[i][j]=True
        dfs(i,j,1,board[i][j])
        visited[i][j]=False

print(answer)
```

# 퇴사
https://www.acmicpc.net/problem/14501

```python
n = int(input())
board = []
for i in range(n):
    t,p = map(int,input().split())
    board.append([t,p])
# 이전값을 활용할 배열 생성
d=[0]*(n+1)

max_val=0
#거꾸로 확인
for i in range(n-1,-1,-1):
    #일을 햇을 때 끝나는 시간
    time=i+board[i][0]
    #끝나는 시간이 그 해당일을 넘어가지 않는다면
    if time<=n:
        # 그 이전 것 값과 현재 값을 더한 것과, 이전까지 값중 가장 큰 값 비교하여 선택
        d[i]=max(d[time]+board[i][1],max_val)
        max_val=d[i]
    # 끝나는 시간이 해당일을 초기화 할 경우 그때까지 최고 금액 변환 X
    else:
        d[i]=max_val
print(max_val)
```

# 연구소
https://www.acmicpc.net/problem/14502

```python
import copy
from collections import deque
n,m = map(int,input().split())
board = []
possible = []
virus = deque()

for i in range(n):
    data = list(map(int,input().split()))
    for j,d in enumerate(data):
        if d==0:
            possible.append([i,j])
        elif d==2:
            virus.append([i,j])
    board.append(data)

if len(possible)<3:
    print(-1)
    exit(0)

dx = [1,-1,0,0]
dy = [0,0,1,-1]

def bfs(virus,board):
    while virus:
        x,y = virus.popleft()
        for i in range(4):
            nx,ny = x+dx[i], y+dy[i]
            if 0<=nx<n and 0<=ny<m and board[nx][ny]==0:
                board[nx][ny] = 2
                virus.append([nx,ny])

def get_count(board):
    cnt = 0
    for x in range(n):
        for y in range(n):
            if board[x][y]==0:
                cnt+=1
    return cnt

answer = 0
def dfs(depth, tmp):
    global answer
    if len(tmp)>=3:
        graph = copy.deepcopy(board)
        tmp_virus =copy.deepcopy(virus)
        for x,y in tmp:
            graph[x][y] = 1
        bfs(tmp_virus, graph)
        answer = max(answer, get_count(graph))
        return
    for i in range(depth, len(possible)):
        tmp.append(possible[i])
        dfs(i+1,tmp)
        tmp.pop()

dfs(0,[])
print(answer)
```

# 로봇청소기
https://www.acmicpc.net/problem/14503

```python
from collections import deque

n,m = map(int,input().split())
sx,sy,sd = map(int,input().split())
# 북 : 0, 동 : 1, 남 : 2, 서 : 3
graph = []
for _ in range(n):
    data = list(map(int,input().split()))
    graph.append(data)
visited = [[0]*m for _ in range(n)]
dx = [-1,0,1,0]
dy = [0,1,0,-1]

def turn(d):
    return (d-1)%4
answer = 1

def bfs(cx,cy,cd):
    global answer
    q = deque()
    q.append([cx,cy,cd])
    visited[cx][cy] = 1
    while q:
        x,y,d = q.popleft()
        _possible = False
        for i in range(4):
            d = turn(d)
            nx,ny = x+dx[d], y+dy[d]
            if 0<=nx<n and 0<=ny<m and graph[nx][ny] == 0 and visited[nx][ny] == 0:
                answer +=1
                visited[nx][ny] = 1
                q.append([nx,ny,d])
                _possible = True
                break
        if not _possible:
            nx, ny = x - dx[d], y - dy[d]
            if graph[nx][ny] == 1:
                break
            else:
                q.append([nx,ny,d])
bfs(sx,sy,sd)

print(answer)
```

# 연산자 끼워넣기
https://www.acmicpc.net/problem/14888

```python
n = int(input())
array = list(map(int,input().split()))
calc = list(map(int,input().split()))
k = sum(calc)
calc_lst = []
for idx,d in enumerate(calc):
    calc_lst.extend([idx]*d)
print(calc_lst)
visited = [False] * n
answer1 = -int(1e9)
answer2 = int(1e9)

def dfs(depth,tmp):
    global answer1, answer2
    if depth>=k:
        answer1 = max(answer1,calculation(array,tmp))
        answer2 = min(answer2,calculation(array,tmp))
        return
    for i in range(k):
        if not visited[i]:
            visited[i] = True
            tmp.append(calc_lst[i])
            dfs(depth+1,tmp)
            tmp.pop()
            visited[i] = False

def calculation(array,tmp):
    answer = array[0]
    for i in range(k):
        if tmp[i]==0:
            answer +=array[i+1]
        elif tmp[i]==1:
            answer -=array[i+1]
        elif tmp[i]==2:
            answer *=array[i+1]
        else:
            answer =int(answer/array[i+1])
    return answer

dfs(0,[])
print(answer1)
print(answer2)
```

# 스타트링크
https://www.acmicpc.net/problem/14889

```python
n = int(input())
board = []
for _ in range(n):
    data = list(map(int,input().split()))
    board.append(data)

visited = [False] * n
answer = int(1e9)
def dfs(depth,idx):
    global answer
    if depth >= n//2:
        answer1,answer2=0,0
        for x in range(n):
            for y in range(n):
                if visited[x] and visited[y]:
                    answer1+=board[x][y]
                elif not visited[x] and not visited[y]:
                    answer2+=board[x][y]
        answer = min(answer, abs(answer1-answer2))
        return

    for i in range(idx, n):
        if not visited[i]:
            visited[i] = True
            dfs(depth+1,i+1)
            visited[i] = False

dfs(0,0)
print(answer)
```
