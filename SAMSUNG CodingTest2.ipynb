## 마법사 상어와 파이어볼
https://www.acmicpc.net/problem/20056

```python
n,m,k=map(int,input().split())
graph=[[[]for _ in range(n)] for _ in range(n)]

# 0,1,2,3,4,5,6,7
dx=[-1,-1,0,1,1,1,0,-1]
dy=[0,1,1,1,0,-1,-1,-1]
q=[]
for i in range(m):
    x,y,m,s,d=map(int,input().split())
    q.append([x-1,y-1,m,s,d])
    
#처음 길이
length=len(q)
for i in range(k):
    
    while q:
        x,y,m,s,d=q.pop(0)
        nx,ny=x+dx[d]*s,y+dy[d]*s
        nx,ny=nx%n,ny%n
        graph[nx][ny].append([m,s,d])

    
    #하나씩 확인하여 1개의 파이어볼이 있는지 2개가 있는지 없는지 확인
    for a in range(n):
        for b in range(n):       
            #2개 이상인 경우, 조건 수행
            if len(graph[a][b])>1:
                tm,ts=0,0
                cnt1,cnt2=0,0
                cnt=0
                # 해당위치에 있는 것을 다 빼준다
                while graph[a][b]:
                    pm,ps,pd=graph[a][b].pop(0)
                    tm+=pm
                    ts+=ps
                    if pd%2==0:
                        cnt2+=1
                    else:
                        cnt1+=1
                    cnt+=1
                # 조건에 맞게 구하기
                tm=tm//5
                ts=ts//cnt
                #질량이 0이면 소멸
                if tm==0:
                    continue
                
                for j in range(4):
                    #모두 홀수나 짝수인 경우
                    if cnt1==cnt or cnt2==cnt:
                        q.append([a,b,tm,ts,j*2])
                    else:
                        q.append([a,b,tm,ts,(j*2)+1])

            #1개인경우
            if len(graph[a][b])==1:
                q.append([a,b]+graph[a][b].pop())

result=0
for item in q:
    result+=item[2]
print(result)
```

## 마법 상어와 토네이도(20057)
https://www.acmicpc.net/problem/20057

```python
n=int(input())
graph=[]
for _ in range(n):
    data=list(map(int,input().split()))
    graph.append(data)

# 서 남 동 북
dx=[0,1,0,-1]
dy=[-1,0,1,0]


# 왼쪽 모래 흩뿌리기 a를 제일 마지막에 두어 나머지 다 계산 후 마지막 처리
left = [(1, 1, 0.01), (-1, 1, 0.01), (1, 0, 0.07), (-1, 0, 0.07), (1, -1, 0.1),
         (-1, -1, 0.1), (2, 0, 0.02), (-2, 0, 0.02), (0, -2, 0.05), (0, -1, 0)]
right=[(x,-y,z) for x,y,z in left]
up=[(y,x,z) for x,y,z in left]
down=[(-y,x,z) for x,y,z in left]

#토네이도 방향
dict = {0: left, 1: down, 2: right, 3: up}


# 모래 계산하는 함수
def recount(x, y, direction):
    global result

    if y < 0:
        return
        
    # 모래 흩뿌리기
    # a 구하기 위한 변수
    total = 0  
    # left,down,right,up중 하나
    for dx, dy, z in direction:
        nx = x + dx
        ny = y + dy
         # a(나머지)
        if z == 0: 
            tmp = graph[x][y] - total
        # 모래 흩뿌리는 비율 계싼
        else:  
            tmp = int(graph[x][y] * z)
            total += tmp
            
        # 범위 내이면 값 갱신
        if 0 <= nx < n and 0 <= ny < n:  
            graph[nx][ny] += tmp
        # 범위 밖이면 result 카운트
        else:  
            result += tmp

nx=ny=int(n/2)
direction=0
#길이
length=1
# 두번씩마다 길이 1증가
count=0
#결과
result=0
            
while True:
    
    for i in range(length):
        #소멸
        if nx<=0 and ny<=0:
            break
        nx,ny=nx+dx[direction],ny+dy[direction]
        recount(nx,ny,dict[direction])
    #소멸
    if nx<=0 and ny<=0:
        break

    direction=(direction+1)%4    
    count+=1

    if count==2:
        count=0
        length+=1
    
print(result)
```

## 마법 상어와 파이어스톰(20058)
https://www.acmicpc.net/problem/20058

```python
import copy
from collections import deque
z,q=map(int,input().split())
n=2**z
dx=[1,-1,0,0]
dy=[0,0,1,-1]
graph=[]
for i in range(n):
    data=list(map(int,input().split()))
    graph.append(data)
#인접한거 확인
def bfs(graph):
    minus_group=deque()
    for i in range(n):
        for j in range(n):
            #인접 얼음 개수 세기
            cnt=0
            for k in range(4):
                nx=i+dx[k]
                ny=j+dy[k]
                # 인접한게 가능한 경우와 얼음이 1이상인경우
                if 0<=nx<n and 0<=ny<n and graph[nx][ny]>=1:
                    cnt+=1
            #3개이상 인접한 경우 그 해당 얼음 감소
            if cnt<3 and graph[i][j]>=1:
                minus_group.append([i,j])

    # 무게를 빼야할 얼음이 존재하는 경우
    while minus_group:
        a,b=minus_group.popleft()
        graph[a][b]-=1

# 남아있는 얼음 중 최대 덩어리 구하기
def max_area_total(graph):
    visited=[[False]*n for _ in range(n)]
    #임시로 저장할 곳
    tmp=deque()
    result=0
    total=0

    for i in range(n):
        for j in range(n):
            #해당 칸에 얼음이 있다면
            if graph[i][j]!=0 and not visited[i][j]:
                # 그 해당칸을 true로 만들고 그 위치를 queue에 삽입
                visited[i][j]=True
                tmp.append([i,j])
                
            # 덩어리 사이즈
            max_area=0
            # 해당 위치가 갈 수 있는 곳을 모두 가고 그 크기의 최대값을 모든 좌표를 돌며 구한다
            while tmp:
                a,b=tmp.popleft()
                max_area+=1
                total+=graph[a][b]
                for k in range(4):
                    na,nb=a+dx[k],b+dy[k]
                    #갈 수 있는 곳이라면 간다
                    #갈 수 있는 곳은 다 가니 visited가 true인 겹치는 곳은 없다
                    if 0<=na<n and 0<=nb<n and not visited[na][nb] and graph[na][nb]!=0:
                        # 한 곳을 갈때마다 구역크기 1씩 증가
                        visited[na][nb]=True
                        tmp.append([na,nb])
            
            result=max(max_area,result)
    return result,total

# l: 2**l
def rotate(l,graph):
    board=[[0]*(n) for _ in range(n)]
    start=0
    for a in range(0,n,l):
        for b in range(0,n,l):
            for x in range(l):
                for y in range(l):
                    #90도 회전 각 크기마다 이동하며 그 범위만큼 90도로 회전
                    board[x+a][y+b]=graph[a+l-y-1][x+b]
    return board


move=list(map(int,input().split()))
#새로운 판에서 수행하기 위해
board=copy.deepcopy(graph)
for i in move:
    l=2**i
    board=rotate(l,board)
    bfs(board)

    
count,total=max_area_total(board)
print(total)
print(count)
```

## 상어 초등학교(21608)
https://www.acmicpc.net/problem/21608

```python
from collections import deque
t=int(input())
n=t**2
student=[[] for _ in range(n + 1)]
graph=[[0]*t for _ in range(t)]
start=[]
for i in range(n):
    stu_num,first,sec,thir,fourth=map(int,input().split())
    student[stu_num].extend([first,sec,thir,fourth])
    start.append(stu_num)
    
dx=[1,-1,0,0]
dy=[0,0,1,-1]

def blank(s_num):
    q=[]
    for i in range(t):
        for j in range(t):
            #빈칸 찾기
            if graph[i][j]==0:
                q.append([i,j])
                
    #1번 조건
    # 인접한 좋아하는 사람 수
    max_tmp=-1
    love_tmp=[]
    for x,y in q:
        love_people=0
        for i in range(4):
            nx=x+dx[i]
            ny=y+dy[i]
            if 0<=nx<t and 0<=ny<t:
                if graph[nx][ny] in student[s_num]:
                    love_people+=1
        # 좋아하는 사람이 인접한 곳에 가장많은 좌표
        if max_tmp<=love_people:
            max_tmp=love_people
            love_tmp.append([max_tmp,x,y])
    # 최대값이 여러개 존재 가능
    love=[[x,y] for m,x,y in love_tmp if m==max_tmp]
    #2번 조건
    # 인접한 비어있는 칸

    if len(love)>1:
        max_tmp=-1
        blank_tmp=[]
        for x,y in love:
            blank_num=0
            for i in range(4):
                nx=x+dx[i]
                ny=y+dy[i]
                if 0<=nx<t and 0<=ny<t:
                    if graph[nx][ny]==0:
                        blank_num+=1
            
            if max_tmp<=blank_num:
                max_tmp=blank_num
                blank_tmp.append([max_tmp,x,y])
        # 최대값이 여러개 존재 가능
        blank_arr=[[x,y] for m,x,y in blank_tmp if m==max_tmp]
    else:
        x,y=love.pop(0)
        graph[x][y]=s_num
        return
    
    #3번 조건
    # 행이 작은 순, 열이 작은 순
    if len(blank_arr)>1:
        # 행과 열 순으로 정렬 후 가장 앞에 있는 것 선택
        last=sorted(blank_arr, key=lambda x:(x[0],x[1]))
    else:
        x,y=blank_arr.pop(0)
        graph[x][y]=s_num
        return
    x,y=last.pop(0)
    graph[x][y]=s_num
    return

#학생의 만족도 구하기
def greet():
    global result
    
    for i in range(t):
        for j in range(t):
            tmp=[]
            student_number=graph[i][j]
            cnt=0
            # 인접한 구역에 좋아하는 사람이 있는지 확인, 좋아하는 사람 수 세기
            for k in range(4):
                nx=i+dx[k]
                ny=j+dy[k]
                if 0<=nx<t and 0<=ny<t:
                    if graph[nx][ny] in student[student_number]:
                        cnt+=1
            if cnt==0:
                continue
            else:
                result=result+(10**(cnt-1))
            
            

for s_num in start:
    blank(s_num)

result=0
greet()
print(result)
```

```python
from collections import deque
n,m=map(int,input().split())
graph=[]
for i in range(n):
    data=list(map(int,input().split()))
    graph.append(data)
#사라진 영역 값 : -2


#반시계 90도로 회전
def rotate(graph):
    board=[[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            board[i][j]=graph[j][n-i-1]
            
    return board




# #중력
def gravity(graph):
    for i in range(n-1,-1,-1):
        for j in range(n):
            prev=graph[i][j]
            if prev>-1:
                tmp=i
                #갈 수 있는 곳으로 내려가기
                while True:
                    if 0<=tmp+1<n and graph[tmp+1][j]==-2:
                        graph[tmp+1][j]=graph[tmp][j]
                        graph[tmp][j]=-2
                        tmp+=1
                    else:
                        break


dx=[1,-1,0,0]
dy=[0,0,1,-1]

#bfs 찾기
def bfs(i,j,tmp):
    q=deque()
    q.append([i,j])
    #노말과 레인보우 담을 배열
    normal_blocks=[[i,j]]
    rainbow_blocks=[]
    #노말과 레인보우 개수
    normal_cnt,rainbow_cnt=1,0
    
    while q:
        x,y=q.popleft()
        for k in range(4):
            nx,ny=x+dx[k],y+dy[k]
            if 0<=nx<n and 0<=ny<n:
                #일반 블록인 경우
                if not visited[nx][ny] and graph[nx][ny]==tmp:
                    visited[nx][ny]=1
                    q.append([nx,ny])
                    normal_cnt+=1
                    normal_blocks.append([nx,ny])
                #무지개인경우
                elif not visited[nx][ny] and graph[nx][ny]==0:
                    visited[nx][ny]=1
                    q.append([nx,ny])
                    rainbow_cnt+=1
                    normal_cnt+=1
                    rainbow_blocks.append([nx,ny])
                    
    #무지개는 다시 갈 수 있으니 방문 해제
    for x,y in rainbow_blocks:
        visited[x][y]=0
                    
    return [normal_cnt,rainbow_cnt,normal_blocks+rainbow_blocks]

#해당 그룹 삭제
def remove(array):
    for x,y in array:
        graph[x][y]=-2

score=0

while True:
    #방문 체크
    visited=[[0]*n for _ in range(n)]
    # 블록 집합
    blocks=[]
    
    
    #오토플레이
    #1.
    for i in range(n):
        for j in range(n):
            if graph[i][j]>0 and not visited[i][j]:
                visited[i][j]=1
                block_tmp=bfs(i,j,graph[i][j])
                #일반 블록이 1개이상있어야한다
                if block_tmp[0]>=2:
                    blocks.append(block_tmp)
                    
    blocks.sort(reverse=True)
    #블록집하이 없는 경우
    if not blocks:
        break
    remove(blocks[0][2])
    score+=(blocks[0][0])**2

    #3번
    gravity(graph)
    #4번
    graph=rotate(graph)
    gravity(graph)
print(score)
```

## 상어 비바라기(21610)
https://www.acmicpc.net/problem/21610

```python
from collections import deque
n,m=map(int,input().split())

graph=[]
for i in range(n):
    data=list(map(int,input().split()))
    graph.append(data)

    
# 1 2 3 4 5 6 7 8
dx=[0,-1,-1,-1,0,1,1,1]
dy=[-1,-1,0,1,1,1,0,-1]

start=deque()
start.append([n-1,0])
start.append([n-1,1])
start.append([n-2,0])
start.append([n-2,1])

def bfs(d,s):
    q=deque()
    while start:
        x,y=start.popleft()
        #1
        nx,ny=(x+dx[d-1]*s)%n,(y+dy[d-1]*s)%n
        #2
        graph[nx][ny]+=1
        # 구름 이동한 위치는 구름 생성 불가이니 표시
        visited[nx][ny]=True
        q.append([nx,ny])
    return q

#4번 물복사
def water_copy(q):
    while q:
        x,y=q.popleft()
        cnt=0
        for i in range(4):
            #대각선 이동
            nx=x+dx[(i*2)+1]
            ny=y+dy[(i*2)+1]
            if 0<=nx<n and 0<=ny<n:
                # 이동한 칸에 물이 있는 경우 바구니 수 증가
                if graph[nx][ny]>0:
                    cnt+=1
        #바구니 수만큼 물 증가
        graph[x][y]+=cnt
        
#구름 만들기
def make_cloud():
    q=deque()
    for i in range(n):
        for j in range(n):
            # 바구니가 2이상인 곳이며 이전에 구름이 없었던 곳은 구름만들기 가능
            if graph[i][j]>=2 and not visited[i][j]:
                q.append([i,j])
                graph[i][j]-=2
    return q


# 각 수행
for i in range(m):
    d,s=map(int,input().split())
    visited=[[False]*n for _ in range(n)]
    tmp=bfs(d,s)
    water_copy(tmp)
    start=make_cloud()
    
result=sum(map(sum,graph))
print(result)
```

```python
n,m=map(int,input().split())

graph=[]
for i in range(n):
    data=list(map(int,input().split()))
    graph.append(data)
    
    
#폭발한 1,2,3번 구슬
result1=0
result2=0
result3=0
    
dx=[-1,1,0,0]
dy=[0,0,-1,1]

dx2=[0,1,0,-1]
dy2=[-1,0,1,0]

center_x=center_y=int((n+1)/2)-1

#한개의 배열로 만들기
def make_array():
    array=[]
    nx,ny=center_x,center_y
    dire=0
    #거리
    s=1
    #2개 될때마다 거리 증가
    cnt=0
    while True:
        if nx<=0 and ny<=0:
            return array
        for i in range(s):
            
            if nx<=0 and ny<=0:
                return array
            
            nx=nx+dx2[dire]
            ny=ny+dy2[dire]
            if graph[nx][ny]!=0:
                array.append(graph[nx][ny])
        cnt+=1
        dire=(dire+1)%4
        if cnt==2:
            s+=1
            cnt=0

#배열로 다시 그래프 만들기
def make_graph(arr):
    nx,ny=center_x,center_y
    dire=0
    s=1
    cnt=0
    idx=0
    while True:
        if nx<=0 and ny<=0:
            return graph
        
        for i in range(s):
            
            if nx<=0 and ny<=0:
                return graph
            
            nx=nx+dx2[dire]
            ny=ny+dy2[dire]
            if idx<=len(arr):
                graph[nx][ny]=arr[idx]
                idx+=1
            else:
                graph[nx][ny]=0
        
        cnt+=1
        dire=(dire+1)%4
        if cnt==2:
            s+=1
            cnt=0

#투척하여 없애기
def throw(d,s):
    nx,ny=center_x,center_y
    for i in range(s):
        nx=nx+dx[d-1]
        ny=ny+dy[d-1]
        graph[nx][ny]=0

#return 크기가 0일때까지 반복
#연속되는 수가 있는지 확인
def check_continue(arr):
    del_arr=[]
    cnt=0
    length=len(arr)
    start,end=0,1
    #투포인터 사용
    while start<length:
        end=start+1
            
        while end<length:
            # 같은 것일 경우 패스
            if arr[start]==arr[end]:
                end+=1
                continue
            #투 포인터가 가르키는 값이 다를 경우 탈출
            else:
                break
        num=end-start
        if num>=4:
            for i in range(start,end):
                del_arr.append(i)
        
        start=end
        
    return del_arr

#같은 숫자들 삭제
def explosion(arr,delete):
    global result1,result2,result3
    board=[]
    for i in range(len(arr)):
        if i in delete:
            continue
        board.append(arr[i])
        
    for i in delete:
        cnt=arr[i]
        if cnt==1:
            result1+=1
        elif cnt==2:
            result2+=2
        elif cnt==3:
            result3+=3
    return board

#구슬 추가
def add_arr(arr):
    board=[]
    cnt=0
    length=len(arr)
    start,end=0,1
    #투포인터 사용
    while start<length:
        end=start+1
        #가능한 용량을 넘어선 경우
        if len(board)>n**2:
            break
            
        while end<length:
            # 같은 것일 경우 패스
            if arr[start]==arr[end]:
                end+=1
                continue
            #투 포인터가 가르키는 값이 다를 경우 탈출
            else:
                break
            
        
        num=end-start
        # 앞 뒤가 서로 다른경우
        if num==1:
            board.append(1)
            board.append(arr[start])
        #같은 것이 2개인 경우
        elif num==2:
            board.append(2)
            board.append(arr[start])
        #같은 것이 3개인 경우
        elif num==3:
            board.append(3)
            board.append(arr[start])
        #end자리로 start 이동
        start=end
    
    #부족한 경우에는 0으로 채워넣기
    while len(board)<=n**2:
        board.append(0)
    
    while len(board)>n**2:
        board.pop()
        
    return board
            
for i in range(m):
    d,s=map(int,input().split())
    throw(d,s)
    arr=make_array()
    while True:
        delete=check_continue(arr)
        if len(delete)==0:
            break
        #같은 구슬 폭발
        arr=explosion(arr,delete)
    #구슬 추가
    arr=add_arr(arr)
    graph=make_graph(arr)
result=result1+result2+result3
print(result)
```

## 주사위 굴리기2(23288)

```python
from collections import deque
n,m,k=map(int,input().split())

#주사위 회전
def rotate(direct):
    a,b,c,d,e,f=array[0],array[1],array[2],array[3],array[4],array[5]
    #6,5,4,3,2,1
    # 동
    if direct==0:
        #3,5,6,1,2,4
        array[0],array[1],array[2],array[3],array[4],array[5]=d,b,a,f,e,c
    #남
    elif direct==1:
        #2,6,4,3,1,5
        array[0],array[1],array[2],array[3],array[4],array[5]=e,a,c,d,f,b
    #서
    elif direct==2:
        #4,5,1,6,2,3
        array[0],array[1],array[2],array[3],array[4],array[5]=c,b,f,a,e,d
    #북
    else:
        #5,1,4,3,6,2
        array[0],array[1],array[2],array[3],array[4],array[5]=b,f,c,d,a,e
#동 : 0, 남: 1, 서: 2, 북: 3
dx=[0,1,0,-1]
dy=[1,0,-1,0]

#연속해서 갈 수 있는 곳(이동한 위치)
def bfs(cx,cy):
    q=deque()
    q.append([cx,cy])
    visited[cx][cy]=True
    possible=1
    while q:
        x,y=q.popleft()
        for i in range(4):
            nx=x+dx[i]
            ny=y+dy[i]
            if 0<=nx<n and 0<=ny<m:
                if not visited[nx][ny] and graph[nx][ny]==graph[cx][cy]:
                    visited[nx][ny]=True
                    possible+=1
                    q.append([nx,ny])
    return possible

#주사위 이동(현재위치,이동방향)
def move(cx,cy,direct):
    nx,ny=cx+dx[direct],cy+dy[direct]
    return nx,ny

#현재 위치와 연속해서 이동할 수 있는 칸 수(이동가능한 칸수,현재 위치)
def score(possible,cx,cy):
    s=possible*graph[cx][cy]
    return s

#이동 후 다음 이동 방향 결정(주사위 배열,현재위치,현재방향)
def decision_direct(arr,cx,cy,direct):
    a=arr[0]
    b=graph[cx][cy]
    
    nx,ny=cx+dx[direct],cy+dy[direct]
    
    if 0>nx or nx>=n or 0>ny or ny>=m:
        direct=(direct+2)%4
    else:
        if a>b:
            direct=(direct+1)%4
        elif a<b:
            direct=(direct-1)%4
    return direct


graph=[]
for i in range(n):
    data=list(map(int,input().split()))
    graph.append(data)

#시작 지점
s_x,s_y=0,0
#동 : 0, 남: 1, 서: 2, 북: 3
s_dir=0
#아랫면부터 주사위
array=[6,5,4,3,2,1]
result=0

for i in range(k):
    s_x,s_y=move(s_x,s_y,s_dir)
    rotate(s_dir)
    visited=[[False]*m for _ in range(n)]
    tmp=bfs(s_x,s_y)
    result+=score(tmp,s_x,s_y)
    s_dir=decision_direct(array,s_x,s_y,s_dir)
    print(s_x,s_y,s_dir,result,array[0],tmp)
print(result)
```

```python
import copy

def move_fish():
    """
    물고기 이동
    1. 상어가 있는 칸, 물고기 냄새 칸, 벗어나는 칸 x 
    2. 45도 반시계 회전 후 이동. 이동 못하는 경우 그대로 
    :return:
    """
    res = [[[] for _ in range(4)] for _ in range(4)]
    for x in range(4):
        for y in range(4):
            while temp[x][y]:
                d = temp[x][y].pop()
                for i in range(d, d - 8, -1):
                    i %= 8
                    nx, ny = x + f_dx[i], y + f_dy[i]
                    if (nx, ny) != shark and 0 <= nx < 4 and 0 <= ny < 4 and not smell[nx][ny]:
                        res[nx][ny].append(i)
                        break
                else:
                    res[x][y].append(d)
    return res

def dfs(x, y, dep, cnt, visit):
    """
    상어 3칸 이동
    1. 제외되는 물고기 수가 많고 > 이동방법 사전순(백트래킹하면 자동으로 됨) 
    2. 이동한 곳을 저장 > 물고기 냄새가 됨  
    """
    global max_eat, shark, eat
    if dep == 3:   # 3번 이동한 경우 그만 
        if max_eat < cnt:
            max_eat = cnt
            shark = (x, y)
            eat = visit[:]
        return
    for d in range(4):
        nx = x + dx[d]
        ny = y + dy[d]
        if 0 <= nx < 4 and 0 <= ny < 4:
            if (nx, ny) not in visit:  # 처음 방문, cnt에 죽은 물고기 수 추가  
                visit.append((nx, ny))
                dfs(nx, ny, dep + 1, cnt + len(temp[nx][ny]), visit)
                visit.pop()
            else:  # 방문한 경우
                dfs(nx, ny, dep + 1, cnt, visit)

#       ←, ↖,   ↑,  ↗, →, ↘, ↓, ↙
f_dx = [0, -1, -1, -1, 0, 1, 1, 1]
f_dy = [-1, -1, 0, 1, 1, 1, 0, -1]
dx = [-1, 0, 1, 0]
dy = [0, -1, 0, 1]

m, s = map(int, input().split())
fish = [list(map(int, input().split())) for _ in range(m)]
graph = [[[] for _ in range(4)] for _ in range(4)]

for x, y, d in fish:
    graph[x - 1][y - 1].append(d - 1)

shark = tuple(map(lambda x: int(x) - 1, input().split()))
smell = [[0] * 4 for _ in range(4)]

for _ in range(s):
    eat = list()
    max_eat = -1
    # 1. 모든 물고기 복제
    temp = copy.deepcopy(graph)
    # 2. 물고기 이동
    temp = move_fish()
    # 3. 상어이동 - 백트래킹
    dfs(shark[0], shark[1],0, 0, list())
    for x, y in eat:
        if temp[x][y]:
            temp[x][y] = []
            smell[x][y] = 3   # 3번 돌아야 없어짐
    # 4. 냄새 사라짐 
    for i in range(4):
        for j in range(4):
            if smell[i][j]:
                smell[i][j] -= 1
    # 5. 복제 마법
    for i in range(4):
        for j in range(4):
            graph[i][j] += temp[i][j]

# 물고기 수 구하기 
answer = 0
for i in range(4):
    for j in range(4):
        answer += len(graph[i][j])

print(answer)
```

## 온풍기 안녕(23289)

```python
from collections import deque
r,c,k=map(int,input().split())
graph=[]
#3가지 옵션을 넣기 위해 사용
#판 0은 현재 온도, 1은 받을 온도, 2는 준 온도
board=[[[0,0,0] for _ in range(c)] for _ in range(r)]
#온풍기
device=[]
#조사해야할 칸
check_list=[]
for i in range(r):
    data=list(map(int,input().split()))
    graph.append(data)
for i in range(r):
    for j in range(c):
        if graph[i][j]==5:
            check_list.append([i,j])
        #온풍기 방향에 따라 설정
        # tmp : 0 위 1우측  2 아래 3 좌측
        elif graph[i][j]>0:
            if graph[i][j]==1:
                tmp=1
            elif graph[i][j]==2:
                tmp=3
            elif graph[i][j]==3:
                tmp=0
            elif graph[i][j]==4:
                tmp=2
            device.append([tmp,i,j])

            


    
w=int(input())
wall=[[[False]*4 for _ in range(c)] for _ in range(r)]
for i in range(w):
    x,y,t=map(int,input().split())
    # t:1이면 (x,y)와(x,y+1)사이 벽,  t:0이면 (x,y)와(x-1,y)사이 벽
    
    # tmp : 0위  1우측  2 아래 3 좌측
    # 0번은 위쪽, 1번은 우측, 2번은 아래쪽, 3번은 좌측
    if t==0:
        wall[x-1][y-1][0]=wall[x-2][y-1][2]=True
    else:
        wall[x-1][y-1][1]=wall[x-1][y][3]=True

# 북,동,남,서
dx=[-1,0,1,0]
dy=[0,1,0,-1]

#바람 불기
def spread():
    for num,x,y in device:
        visited=[[False]*c for _ in range(r)]
        cx,cy=x+dx[num],y+dy[num]
        board[cx][cy][0]+=5
        visited[cx][cy]=True
        q=deque()
        q.append([cx,cy])
        
        for size in range(4,0,-1):
            if not q:
                break
            length=len(q)
            for _ in range(length):
                #3곳 체크
                x,y=q.popleft()
                #대각선 위쪽
                nx=x+dx[(num)%4]+dx[(num-1)%4]
                ny=y+dy[(num)%4]+dy[(num-1)%4]
                #방문한적 없으며 위쪽과 좌측 벽이 없어야한다
                if 0<=nx<r and 0<=ny<c and not visited[nx][ny] and not wall[nx][ny][(num+2)%4] and not wall[x][y][(num-1)%4]:
                    visited[nx][ny]=True
                    board[nx][ny][0]+=size
                    q.append([nx,ny])

                #가운데
                nx=x+dx[num]
                ny=y+dy[num]
                if 0<=nx<r and 0<=ny<c and not visited[nx][ny] and not wall[nx][ny][(num+2)%4] and not wall[x][y][(num)%4]:
                    visited[nx][ny]=True
                    board[nx][ny][0]+=size
                    q.append([nx,ny])

                #아래 대각선
                nx=x+dx[(num)%4]+dx[(num+1)%4]
                ny=y+dy[(num)%4]+dy[(num+1)%4]
                if 0<=nx<r and 0<=ny<c and not visited[nx][ny] and not wall[nx][ny][(num+2)%4] and not wall[x][y][(num+1)%4]:
                    visited[nx][ny]=True
                    board[nx][ny][0]+=size
                    q.append([nx,ny])
                

def set_temp():
    #온도 전해주기
    for x in range(r):
        for y in range(c):
            #없으면 못전해주기때문
            if board[x][y][0]==0:
                continue
                
            for i in range(4):
                nx=x+dx[i]
                ny=y+dy[i]
                #근처값보다 큰 경우
                if 0<=nx<r and 0<=ny<c and not wall[nx][ny][(i+2)%4] and board[x][y][0]>board[nx][ny][0]:
                    board[nx][ny][1]+=(board[x][y][0]-board[nx][ny][0])//4
                    board[x][y][2]+=(board[x][y][0]-board[nx][ny][0])//4
    #온도 빼주기            
    for x in range(r):
        for y in range(c):
            board[x][y][0]=board[x][y][0]+board[x][y][1]-board[x][y][2]
            board[x][y][1]=board[x][y][2]=0
            
    #가장자리 온도 빼주기
    for x in range(r):
        for y in range(c):
            if x==0 or y==0 or x==r-1 or y==c-1:
                if board[x][y][0]>0:
                    board[x][y][0]-=1

#검사하기
def check():
    for x,y in check_list:
        if board[x][y][0]<k:
            return False
    return True


cnt=0
fail=False
while True:
    spread()
    set_temp()
    cnt+=1
    flag=check()
    
    if flag:
        break
        
    if cnt>100:
        fail=True
        break

if fail:
    print(-1)
else:
    print(cnt)
```

```python
n,k=map(int,input().split())
array=list(map(int,input().split()))

#물고기 넣기
def add_fish(arr):
    min_val=min(arr)
    for i in range(len(arr)):
        if arr[i]==min_val:
            arr[i]+=1
    return arr

#어항 쌓기 90도
def stack_first(arr):
    board=[[] for _ in range(n)]
    
    #현재 물고기수, 들어올 물고기수, 나갈 물고기수
    for i in range(n):
        board[i].append([arr[i],0,0])

    k=0
    data=board.pop(k)
    board[0].append(data[0])
    idx=0
    
    while True:
        for i in range(len(board)):
            if len(board[i])>1:
                idx=i
                
        #쌓는게 불가능할 때, 그 쌓을려는 크기보다 밑의 배열 수가 더 작은 경우
        if len(board)-1-idx<len(board[idx]):
            break
            
        #90도로 어항 쌓기
        for i in range(idx,-1,-1):
            data=board.pop(i)
            for j in range(len(data)):
                board[i+j].append(data[j])
                
    return board


#어항 물고기 수 조정
def change_fish_num(board):
    dx=[1,-1,0,0]
    dy=[0,0,1,-1]
    for y in range(len(board)):
        for x in range(len(board[y])):
            for i in range(4):
                nx=x+dx[i]
                ny=y+dy[i]
                if 0<=nx<len(board[y]) and 0<=ny<len(board) and 0<=ny<len(board[ny]):
                    #차이를 5로 나눈 몫만큼 이동
                    d=(board[y][x][0]-board[ny][nx][0])//5
                    # 0보다 크면 들어올 배열, 나갈 배열에 추가
                    if d>0:
                        #나갈배열
                        board[y][x][2]+=d
                        #들어올 배열
                        board[ny][nx][1]+=d
    
    #물고기 수 처리
    for y in range(len(board)):
        for x in range(len(board[y])):
            board[y][x][0]=board[y][x][0]+(board[y][x][1]-board[y][x][2])
            board[y][x][1]=0
            board[y][x][2]=0
    
    return board

#어항 일렬로 정리
def make_fish_array(board):
    arr=[]
    for y in range(len(board)):
        for x in range(len(board[y])):
            arr.append(board[y][x][0])
    return arr

#180도 회전
def rotate180(arr):
    center=int(n/2)-1
    
    
    #현재 물고기수, 들어올 물고기수, 나갈 물고기수
    board=[[] for _ in range(center+1)]
    reverse_arr=arr[center::-1]
    new_arr=arr[center+1:]
    for i in range(center+1):
        board[i].append(new_arr[i])
        board[i].append(reverse_arr[i])
    #n/4로 나누기
    center=(n//4)-1
    reverse_board=board[center::-1]
    new_board=board[center+1:]
    #새로운 보드 정의
    board=[[] for _ in range(center+1)]
    for i in range(center+1):
        for j in range(len(new_board[i])):
            board[i].append([new_board[i][j],0,0])
        for j in range(len(reverse_board[i])-1,-1,-1):
            board[i].append([reverse_board[i][j],0,0])
    return board

cnt=0
while True:
    array=add_fish(array)
    board=stack_first(array)
    board=change_fish_num(board)
    array=make_fish_array(board)
    board=rotate180(array)
    board=change_fish_num(board)
    array=make_fish_array(board)
    print(array)
    max_val=max(array)
    min_val=min(array)
    cnt+=1
    if max_val-min_val<=k:
        break
    
    
print(cnt)
```
