## 38.정확한 순위
- 선생은 시험을 본 n명의 성적을 분실하고, 성적을 비교한 결과의 일분만 가지고 있다.
- n명의 성적은 모두 다른데, A번 학생의 성적이 B번 학생의 성적보다 낮다면 화살표가 A에서 B로 가리키도록 표현할 수 있다.
- 이 그래프를 통해 학생의 성적 순위를 알 수 있는 학생도 있고 알 수 없는 학생도 있다.
- 성적 순위를 정확히 알 수 있는 학생 수를 구하시오.
- 첫째 줄 학생 수 n, 비교한 횟수 m, 둘째 줄에는 m개의 줄에 A,B의 학생의 성적을 비교한 결과가 주어진다. A가 B보다 성적이 낮다

```python
n,m=map(int,input().split())
INF=int(1e9)
graph=[[INF]*(n+1) for _ in range(n+1)]
for _ in range(m):
    a,b=map(int,input().split())
    graph[a][b]=1
for i in range(1,n+1):
    for j in range(1,n+1):
        if i==j:
            graph[i][j]=0

for k in range(1,n+1):
    for a in range(1,n+1):
        for b in range(1,n+1):
            graph[a][b]=min(graph[a][k]+graph[k][b],graph[a][b])
               
result=0
for a in range(1,n+1):
    cnt=0
    for b in range(1,n+1):
        if graph[a][b]!=INF or graph[b][a]!=INF:
            cnt+=1
    if cnt==n:
        result+=1
print(result)
```

## 39.화상 탐사
- 화성 탐사 기계는 출발 지점에서 목표 지점까지 이동할 때 최적의 경로를 찾아야한다.
- n x n 크기의 2차원 공간에서, 각각의 칸을 지나기 위한 비용이 존재한다. (0,0)위치에서 가장 오른쪽 (n-1,n-1)의 위치로 이동하는 최소 비용을 출력하는 프로그램을 구하시오. 화성 탐사 기계는 특정한 위치에서 상하좌우 인접한 곳으로 1칸씩 이동할 수 있다.
- 첫째줄에는 테스트 케이스 T, 그 후 n이 주어진다. 그 후 각 칸의 비용이 공백으로 주어진다.

```python
import heapq

INF=int(1e9)

t=int(input())

for i in range(t):
    n=int(input())

    graph=[]
    distance=[[INF]*(n) for _ in range(n)]

    for _ in range(n):
        graph.append(list(map(int,input().split())))

    q=[]
    heapq.heappush(q,(graph[0][0],0,0))
    distance[0][0]=graph[0][0]

    dx=[1,-1,0,0]
    dy=[0,0,1,-1]
    while q:
        dis,x,y=heapq.heappop(q)
        if distance[x][y]<dis:
            continue
        for i in range(4):
            nx=x+dx[i]
            ny=y+dy[i]
            if 0<=nx<n and 0<=ny<n:
                cost=dis+graph[nx][ny]
                if cost<distance[nx][ny]:
                    distance[nx][ny]=cost
                    heapq.heappush(q,(cost,nx,ny))

    print(distance[n-1][n-1])
```

## 40. 숨박꼭질
- n개의 헛간 중 동빈이는 1번으로부터 최단거리가 가장 먼 헛간에 숨는다. 술래는 1번 헛간에서 출발한다.
- 총 m개의 양방향 통로가 있다. 동빈이가 숨을 헛간의 번호와 그 헛간까지의 거리, 같은 거리에 있는 헛간의 개수를 출력하시오.

```python
# 다이스트라

import heapq

n,m=map(int,input().split())
INF=int(1e9)

graph=[[] for _ in range(n+1)]
start=1
distance=[INF]*(n+1)

for _ in range(m):
    a,b=map(int,input().split())
    graph[a].append((b,1))
    graph[b].append((a,1))

def dijkstra(start):
    q=[]
    heapq.heappush(q,(0,start))
    distance[start]=0
    while q:
        dis,cur=heapq.heappop(q)
        if distance[cur]<dis:
            continue
        for i in graph[cur]:
            cost=dis+i[1]
            
            if cost<distance[i[0]]:
                distance[i[0]]=cost
                heapq.heappush(q,(cost,i[0]))
dijkstra(start)

max_val=0
for i in range(1,n+1):
    if max_val<distance[i]:
        max_val=distance[i]
        max_index=i

print(max_index,max_val,distance.count(max_val))
            
```

## 41. 여행 계획
- n개의 여행지가 존재, 여행지끼리의 도로가 있으며, 도로는 양방향 통행이 가능하다.
- 한울이는 여행 계획을 세운 뒤에 여행 계획이 가능한지 여부를 판단하고자 한다.
- 첫째 줄에 여행지 수 n, 여행 계획에 속한 도시의 수 m이 주어진다, 둘째 줄에는 n개의 줄에 걸쳐 n X n 행렬로 두 여행지가 서로 연결되어 있는지의 여부가 주어진다. 그 값이 1이면 서로 연결, 0이면 연결X이다. 마지막 줄에는 한울이의 여행 계획에 포함된 여행지의 번호들이 주어진다.
- 여행 계획이 가능하면 YES, 불가능하면 NO를 출력해라.

```python
n,m=map(int,input().split())

# 루트노드 찾기(해당 집합의 루트 찾기)
def find_parent(parent,x):
    if parent[x]!=x:
        parent[x]=find_parent(parent,parent[x])
    return parent[x]

parent=[0]*(n+1)

# 더 작은 값의 집합에 큰 값을 포함
def union_parent(parent,a,b):
    a=find_parent(parent,a)
    b=find_parent(parent,b)
    if a<b:
        parent[b]=a
    else:
        parent[a]=b


# 각 node를 집합으로 생성
for i in range(1,n+1):
    parent[i]=i

graph=[]

# 그래프 입력
for i in range(n):
    graph.append(list(map(int,input().split())))

# print(graph)
# 크루스칼 알고리즘 적용
for a in range(n):
    for b in range(a,n):
        if graph[a][b]==1:
            if find_parent(parent,a+1)!=find_parent(parent,b+1):
                union_parent(parent,a+1,b+1)
                
# print(parent)
plan=list(map(int,input().split()))

result=True
for p in range(len(plan)-1):
    if parent[plan[p]]!=parent[plan[p+1]]:
        result=False

if result:
    print('YES')
else:
    print('NO')
```

## 43.어두인길
- 한 마을은 n개의 집과 m개의 도로로 구성되어 있다. 각 집은 0~n-1번까지의 번호로 구분된다.
- 모든 도로에는 가로등이 구비되어 있는데, 가로등을 하루 동안 켜기 위한 비용은 해당의 도로의 길이와 동일하다.
- 가로등이 켜진 도로만 오갈 수 있다. 이때 일부 가로등을 비활성화하여 최대한 많은 금액을 절약하고자 할 때, 절약할 수 있는 최대 금액을 출력하시오.
- 첫째 줄 집의 개수 n, 도로의 수 m이 주어진다. 둘째 줄에는 m개에 걸쳐 x,y,z(x번 집에서 y번집으로 길이가 z)가 주어진다. 도로는 양방향이다.

```python
n,m=map(int,input().split())

def find_parent(parent,x):
    if parent[x]!=x:
        parent[x]=find_parent(parent,parent[x])
    return parent[x]

def union_parent(parent,a,b):
    a=find_parent(parent,a)
    b=find_parent(parent,b)
    if a<b:
        parent[b]=a
    else:
        parent[a]=b


# 간선 정의
edges=[]
total_cost=0
for i in range(m):
    a,b,c=map(int,input().split())
    edges.append((c,a,b))
    total_cost+=c

# 간선 정렬
edges.sort()

# 각 노드들을 각 집합으로 정의
parent=[0]*(n)
for i in range(n):
    parent[i]=i

result=0

for edge in edges:
    cost,x,y=edge
    if find_parent(parent,x)!=find_parent(parent,y):
        union_parent(parent,x,y)
        result+=cost
        

print(total_cost-result)
```

## 46.아기상어

```python
from collections import deque

n=int(input())

graph=[]
for _ in range(n):
    graph.append(list(map(int,input().split())))
    
dx=[1,-1,0,0]
dy=[0,0,1,-1]

for i in range(n):
    for j in range(n):
        if graph[i][j]==9:
            start=(i,j)
x,y=start
size=2
def bfs(x,y,s):
    distance=[[0]*n for _ in range(n)]
    visited=[[False]*n for _ in range(n)]
    q=deque()
    q.append((x,y))
    visited[x][y]=True
    temp=[]
    while q:
        cur_x,cur_y=q.popleft()
        for i in range(4):
            nx=cur_x+dx[i]
            ny=cur_y+dy[i]
            if 0<=nx<n and 0<=ny<n and visited[nx][ny]==False:
                if graph[nx][ny]<=s:
                    q.append((nx,ny))
                    visited[nx][ny]=True
                    distance[nx][ny]=distance[cur_x][cur_y]+1
                    if graph[nx][ny]<s and distance[nx][ny]!=0:
                        temp.append((nx,ny,distance[nx][ny]))
    temp.sort(key=lambda x:(x[2],x[0],x[1]))
    return temp
    
cnt=0
result=0
while True:
    shark=bfs(x,y,size)
    if len(shark)==0:
        break
    nx,ny,dist=shark.pop(0)
    result+=dist
    graph[x][y],graph[nx][ny]=0,0
    x,y=nx,ny
    cnt+=1
    if cnt==size:
        size+=1
        cnt=0
print(result)
```

```python
# answer
import sys
from collections import deque

input = sys.stdin.readline

# 먹을 수 있는 물고기가 1마리라면, 그 물고기를 먹으러 간다.
# 먹을 수 있는 물고기가 1마리보다 많다면, 거리가 가장 가까운 물고기를 먹으러 간다.

n = int(input())
graph = []
for _ in range(n):
    graph.append(list(map(int,input().split())))


dx = [0,0,1,-1]
dy = [1,-1,0,0]
cnt = 0
x,y,size = 0,0,2
#상어위치
for i in range(n):
    for j in range(n):
        if graph[i][j] == 9:
            x = i
            y = j

def biteFish(x,y,shark_size):
    distance = [[0] * n for _ in range(n)]
    visited = [[0] * n for _ in range(n)]
    # 거리는 아기 상어가 있는 칸에서 물고기가 있는 칸으로 이동할 때, 지나야하는 칸의 개수의 최솟값이다. (bfs사용)
    q = deque()
    q.append((x,y))
    visited[x][y] = 1
    temp = []
    while q:
        cur_x,cur_y = q.popleft()
        for i in range(4):
            nx = cur_x + dx[i]
            ny = cur_y + dy[i]
            if 0<= nx < n and 0<= ny < n and visited[nx][ny] == 0:
                if graph[nx][ny] <= shark_size:
                    q.append((nx,ny))
                    visited[nx][ny] = 1
                    distance[nx][ny] = distance[cur_x][cur_y] + 1
                    if graph[nx][ny] < shark_size and graph[nx][ny] != 0:
                        temp.append((nx,ny,distance[nx][ny]))

# 거리가 가까운 물고기가 많다면, 가장 위에 있는 물고기, 그러한 물고기가 여러마리라면, 가장 왼쪽에 있는 물고기를 먹는다.
    return sorted(temp,key=lambda x: (-x[2],-x[0],-x[1]))


cnt = 0
result = 0
while 1:
    shark = biteFish(x,y,size)
    # 더 이상 먹을 수 있는 물고기가 공간에 없다면 아기 상어는 엄마 상어에게 도움을 요청한다.
    if len(shark) == 0:
        break
    # 거리가 가까운 물고기가 많다면, 가장 위에 있는 물고기, 그러한 물고기가 여러마리라면, 가장 왼쪽에 있는 물고기를 먹는다.
    # 정렬한 결과를 반영해준다.
    nx,ny,dist =shark.pop()
    
    #움직이는 칸수가 곧 시간이 된다.
    result += dist
    graph[x][y],graph[nx][ny] = 0,0
    #상어좌표를 먹은 물고기 좌표로 옮겨준다.
    x,y = nx,ny
    cnt += 1
    if cnt == size:
        size += 1
        cnt = 0
print(result)
```

## 47.청소년상어
https://www.acmicpc.net/problem/19236

```python
import copy

n=4
graph=[[]for _ in range(n)]
for i in range(n):
    data=list(map(int,input().split()))
    for k in range(0,len(data)-1,2):
        graph[i].append([data[k],data[k+1]-1])

# 화살표 이동 방향
dx=[-1,-1,0,1,1,1,0,-1]
dy=[0,-1,-1,-1,0,1,1,1]

def dfs(cur_x,cur_y,score,graph):
    global result
    
    #상어 시작
    #상어 먹이
    score+=graph[cur_x][cur_y][0]
    result=max(result,score)
    graph[cur_x][cur_y][0]=0
    
    
    # 물고기 이동
    for i in range(1,17):
        f_x,f_y=-1,-1
        #1번부터 17번까지의 물고기 위치 찾기
        for x in range(4):
            for y in range(4):
                if graph[x][y][0]==i:
                    f_x,f_y=x,y
                    break
        if f_x==-1 and f_y==-1:
            continue
        #물고기의 방향 얻기
        fish_dir=graph[f_x][f_y][1]
        
        #물고기 이동
        for k in range(8):
            # 여덟방향에 대해 고려
            nd=(fish_dir+i)%8
            nx=f_x+dx[nd]
            ny=f_y+dy[nd]
            #이동 못하는 경우 
            if not(0<=nx<n and 0<=ny<n) or (nx==cur_x and ny==cur_y):
                continue
            # 가능한 경우 서로 교환 후 for문 break
            graph[f_x][f_y][1]=nd
            graph[f_x][f_y],graph[nx][ny]=graph[nx][ny],graph[f_x][f_y]
            break
    
    #상어가 먹기 시작
    shark_dir=graph[cur_x][cur_y][1]
    for i in range(1,5):
        nx=cur_x+dx[shark_dir]*i
        ny=cur_y+dy[shark_dir]*i
        if 0<=nx<n and 0<=ny<n and graph[nx][ny][0]>0:
            dfs(nx,ny,score,copy.deepcopy(graph))
            

result=0
dfs(0,0,0,graph)
print(result)

```

```python
from collections import deque

n,m,k=map(int,input().split())
graph=[]
for _ in range(n):
    graph.append(list(map(int,input().split())))


#새로운 그래프에 상어 번호와 채취 기록
board=[[[0,0]]*n for _ in range(n)]
            
    
# 1: 위, 2: 아래, 3: 왼쪽, 4: 오른쪽
cur_dir=list(map(int,input().split()))

priority=[[]  for _ in range(m)]
for i in range(m):
    for j in range(4):
        priority[i].append(list(map(int,input().split())))

dx=[-1,1,0,0]
dy=[0,0,-1,1]

def update_smell():
    for i in range(n):
        for j in range(n):
            if board[i][j][1]>0:
                board[i][j][1]-=1
            if graph[i][j]!=0:
                board[i][j]=[graph[i][j],k]

                
def move():
    array=[[0]*n for _ in range(n)]
    for x in range(n):
        for y in range(n):
            if graph[x][y]!=0:
                s_dir=cur_dir[graph[x][y]-1]
                found=False
                move_dir=priority[graph[x][y]-1][s_dir-1]
                for i in range(4):
                    nx=x+dx[move_dir[i]-1]
                    ny=y+dy[move_dir[i]-1]
                    if 0<=nx<n and 0<=ny<n:
                        if board[nx][ny][1]==0:
                            s_dir=move_dir[i]
                            if array[nx][ny]==0:
                                array[nx][ny]=graph[x][y]
                            else:
                                array[nx][ny]=min(array[nx][ny],graph[x][y])
                            
                            found=True
                            break
                            
                if found:
                    continue

                for i in range(4):
                    nx=x+dx[move_dir[i]-1]
                    ny=y+dy[move_dir[i]-1]
                    if 0<=nx<n and 0<=ny<n:
                        if board[nx][ny][0]==graph[x][y]:
                            s_dir=move_dir[i]
                            array[nx][ny]=graph[x][y]
                            break
        return array


time=0
while True:
    update_smell()
    new_array=move()
    graph=new_array
    time+=1
    
    check=True
    for i in range(n):
        for j in range(n):
            if graph[i][j]>1:
                check=False
    if check:
        print(time)
        break
    if time>=1000:
        print(-1)
        break
```
